## Data Visulization 

```python
plt.figure(figsize = (14,7))
sns.scatterplot(data = games, x = "Genre", y="Global_Sales")
plt.title("Graph Between Global Sales and Genre")
plt.xlabel("Genre")
plt.ylabel("Global Sales")
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (14,7))
sns.countplot(data = games, x = "Genre")
plt.title("Graph to display the frequency of different Genre")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (14,7))
sns.countplot(data = games, x = "Platform")
plt.title("Graph to display the frequency of different Platforms")
plt.xlabel("Platforms")
plt.ylabel("Count")
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (14,7))
sns.countplot(data = games, x = "Year")
plt.title("Graph to display the number of games released each year")
plt.xlabel("Year")
plt.ylabel("Number of Games")
plt.show()
```

```python
most_games = games.Year.value_counts().head(10)
plt.figure(figsize = (14,7))
sns.barplot(x = most_games.index, y = most_games)
plt.title("Graph to display the top 10 years with most games released")
plt.xlabel("Year")
plt.ylabel("Number of Games")
plt.grid()
plt.show()
```

```python
most_games = games.Genre.value_counts().head(10)
plt.figure(figsize = (14,7))
sns.barplot(x = most_games.index, y = most_games)
plt.title("Graph to display the top 10 Genre with most games released")
plt.xlabel("Genre")
plt.ylabel("Number of Games")
plt.grid()
plt.show()
```

```python
most_games = games.Publisher.value_counts().head()
plt.figure(figsize = (14,7))
sns.barplot(x = most_games.index, y = most_games)
plt.title("Graph to display the top 5 Publishers with most games released")
plt.xlabel("Publisher")
plt.ylabel("Number of Games")
plt.grid()
plt.show()
```
